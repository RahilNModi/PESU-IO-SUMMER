```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.load("LinRegdata_X.npy")
Y=np.load("LinRegdata_Y.npy")
```

```python
plt.scatter(X,Y,color='black')
plt.show()
```

```python
r=LinearRegression().fit(X,Y)
```

```python
r.score(X,Y)
```

```python
print("m: ",r.coef_,"\nb: ",r.intercept_)
```

```python
x=np.linspace(-100,100).reshape(-1,1)
```

```python
plt.scatter(X,Y,color="red")
plt.plot(x,r.predict(x),color='blue')
plt.show()
```

```python
r.predict([[0]])
```

```python
test=[[22],[-55],[99]]
```

```python
plt.scatter(X,Y,color="red")
plt.plot(x,r.predict(x),color='blue')
plt.scatter(test,r.predict(test),color='black')
plt.show()
```
